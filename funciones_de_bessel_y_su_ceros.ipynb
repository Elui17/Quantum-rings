```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn, jn_zeros

# Definir el rango de x
x = np.linspace(0, 20, 1000)

# Crear la grfica
plt.figure(figsize=(10, 6))
for m in range(4):  # Graficar las funciones de Bessel de orden 0 a 3
    y = jn(m, x)
    plt.plot(x, y, label=f'$J_{m}(x)$')

    # Encontrar los primeros ceros de la funcin de Bessel
    zeros = jn_zeros(m, 3)  # Los primeros 3 ceros de la funcin J_m(x)
    plt.scatter(zeros, np.zeros_like(zeros), color='red', zorder=5)  # Sealar los ceros en la grfica

    # Anotar los ceros en la grfica
    for i, zero in enumerate(zeros):
        plt.annotate(f'{zero:.2f}', (zero, 0), textcoords="offset points", xytext=(0,10), ha='center')

# Configurar la grfica
plt.title("Funciones de Bessel de primer tipo con ceros destacados")
plt.xlabel('$x$')
plt.ylabel('$J_m(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
